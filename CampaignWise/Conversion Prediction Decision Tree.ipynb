### Clean the data set

```python
#Checking how many ' ?' in every columns (null values given as ' ?')
data.isin(['--']).sum()
```

```python
#replace all ? as null value
null_data = data.replace(" ?",np.NaN)

#check wheter all ? changed as null
print("\n----checking all ? data changed as null----")
print(null_data.isnull().sum())

#drop the null value rows
notNull_data=null_data.dropna()
print("\n---checking whether all null data cleared----")
print(notNull_data.isnull().sum())
```

```python
#checking the duplicate data entry
print("Duplicate Entries : ",notNull_data.duplicated().sum())

#drop all duplicated entry
clean_data=notNull_data.drop_duplicates()
print("\n---Check whether there are duplicate data present after the clean---- ")
print("\nDuplicate data : ",clean_data.duplicated().sum())
```

# Data Set Pre Process

```python
clean_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

#creating an instance of labelencoder
labelencoder=LabelEncoder()

#assign numerical value and storing it in another
encoded_data=clean_data.iloc[0:]

encoded_data["bidding_strategy_type"]=labelencoder.fit_transform(clean_data["bidding_strategy_type"])
encoded_data["campaign name"]=labelencoder.fit_transform(clean_data["campaign name"])
encoded_data["date"]=labelencoder.fit_transform(clean_data["date"])
```

```python
#drop the Day column
test_data=encoded_data.drop(['cost-per-acquisition (CPA)','conversion per euro','cost_per_conversion'],axis=1)

x=test_data.drop('conversions',axis=1)
x.head()
```

```python
#seperate income column as y
y=test_data['conversions']
print(y)
```

### Split train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print("Training Set of Feature : ",x_train.shape)
print("Test Set of Features : ",x_test.shape)
print("Training set of Target : ",y_train.shape)
print("Test set of Target : ",y_test.shape)
x_train.head()
```

## Prediction

```python
from sklearn.tree import DecisionTreeRegressor
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_3 = DecisionTreeRegressor(max_depth=8)

regr_1.fit(x_train, y_train)
regr_2.fit(x_train, y_train)
regr_3.fit(x_train, y_train)

y_1 = regr_1.predict(x_test)
y_2 = regr_2.predict(x_test)
y_3 = regr_3.predict(x_test)

print("Predicted Value by the model1\n",y_1)
print("Predicted Value by the model2\n",y_2)
print("Predicted Value by the model2\n",y_3)
```

```python
compareResult=pd.DataFrame()
compareResult['Actual']=y_test
compareResult['y_1']=y_1
compareResult['y_2']=y_2
compareResult['y_3']=y_3
```

```python
def get_rules(tree, feature_names, class_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    paths = []
    path = []
    
    def recurse(node, path, paths):
        
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            p1, p2 = list(path), list(path)
            p1 += [f"({name} <= {np.round(threshold, 3)})"]
            recurse(tree_.children_left[node], p1, paths)
            p2 += [f"({name} > {np.round(threshold, 3)})"]
            recurse(tree_.children_right[node], p2, paths)
        else:
            path += [(tree_.value[node], tree_.n_node_samples[node])]
            paths += [path]
            
    recurse(0, path, paths)

    # sort by samples count
    samples_count = [p[-1][1] for p in paths]
    ii = list(np.argsort(samples_count))
    paths = [paths[i] for i in reversed(ii)]
    
    rules = []
    for path in paths:
        rule = "if "
        
        for p in path[:-1]:
            if rule != "if ":
                rule += " and "
            rule += str(p)
        rule += " then "
        if class_names is None:
            rule += "response: "+str(np.round(path[-1][0][0][0],3))
        else:
            classes = path[-1][0][0]
            l = np.argmax(classes)
            rule += f"class: {class_names[l]} (proba: {np.round(100.0*classes[l]/np.sum(classes),2)}%)"
        rule += f" | based on {path[-1][1]:,} samples"
        rules += [rule]
        
    return rules
```
