# Clicks Trend

```python
plt.figure(figsize = (15, 6))
plt.plot(tr_df.index, tr_df['clicks'])
plt.xlabel('Date', {'fontsize': 12})
plt.ylabel('clicks', {'fontsize': 12})
plt.title('Number of clicks')
plt.show()
```

```python
n_cols = 1
clickdataset = tr_df['clicks']
clickdataset = pd.DataFrame(clickdataset)
clickdata = clickdataset.values

clickdata.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range= (0, 1))
click_scaled_data = scaler.fit_transform(np.array(clickdata))
```

```python
train_size = int(len(clickdata) * 0.80)
test_size = len(clickdata) - train_size
print("Train Size :",train_size,"Test Size :",test_size)
```

```python
train_data = click_scaled_data[0:train_size, :]
train_data.shape
```

```python
# Creating a Training set with 60 time-steps and 1 target
x_train = []
y_train = []
time_steps = 60
n_cols = 1

for i in range(time_steps, len(train_data)):
    x_train.append(train_data[i-time_steps:i, :n_cols])
    y_train.append(train_data[i, :n_cols])
    if i<=time_steps:
        print('X_train: ', x_train)
        print('y_train:' , y_train)
```

```python
len(x_train)
```

```python
# Convert to numpy array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the input to (n_samples, time_steps, n_feature)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], n_cols))
```

```python
x_train.shape , y_train.shape
```

```python
model = Sequential([
    LSTM(50, return_sequences= True, input_shape= (x_train.shape[1], n_cols)),
    LSTM(64, return_sequences= False),
    Dense(32),
    Dense(16),
    Dense(n_cols)
])

model.compile(optimizer= 'adam', loss= 'mse' , metrics=['mean_absolute_error'])
```

```python
model.summary()
```

```python
# Fitting the LSTM to the Training set
callbacks = [EarlyStopping(monitor= 'loss', patience= 10 , restore_best_weights= True)]
history = model.fit(x_train, y_train, epochs= 100, batch_size= 32 , callbacks= callbacks )
```

```python
plt.figure(figsize=(12, 8))
plt.plot(history.history["loss"])
plt.plot(history.history["mean_absolute_error"])
plt.legend(['Mean Squared Error','Mean Absolute Error'])
plt.title("Losses")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
# Creating a testing set with 60 time-steps and 1 output
time_steps = 60
test_data = click_scaled_data[train_size - time_steps:, :]

x_test = []
y_test = []
n_cols = 1

for i in range(time_steps, len(test_data)):
    x_test.append(test_data[i-time_steps:i, 0:n_cols])
    y_test.append(test_data[i, 0:n_cols])
x_test, y_test = np.array(x_test), np.array(y_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], n_cols))
```

```python
x_test.shape , y_test.shape
```

```python
predictions = model.predict(x_test)
```

```python
#inverse predictions scaling
predictions = scaler.inverse_transform(predictions)
predictions.shape
```

```python
#inverse y_test scaling
y_test = scaler.inverse_transform(y_test)

RMSE = np.sqrt(np.mean( y_test - predictions )**2).round(2)
RMSE
```

```python
preds_acts = pd.DataFrame(data={'Predicted Clicks':predictions.flatten(), 'Actual Clicks':y_test.flatten()})
preds_acts
```

```python
train = clickdataset.iloc[:train_size , 0:1]
test = clickdataset.iloc[train_size: , 0:1]
test['Predicted Clicks'] = predictions

plt.figure(figsize= (16, 6))
plt.title('Clicks Prediction', fontsize= 18)
plt.xlabel('Date', fontsize= 18)
plt.ylabel('Clicks', fontsize= 18)
plt.plot(train['clicks'], linewidth= 3)
plt.plot(test['clicks'], linewidth= 3)
plt.plot(test["Predicted Clicks"], linewidth= 3)
plt.legend(['Train', 'Test', 'Predictions'])
```
