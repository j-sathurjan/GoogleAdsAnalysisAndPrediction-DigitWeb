### Clean the data set

```python
#Checking how many ' ?' in every columns (null values given as ' ?')
data.isin(['--']).sum()
```

```python
#replace all ? as null value
null_data = data.replace(" ?",np.NaN)

#check wheter all ? changed as null
print("\n----checking all ? data changed as null----")
print(null_data.isnull().sum())

#drop the null value rows
notNull_data=null_data.dropna()
print("\n---checking whether all null data cleared----")
print(notNull_data.isnull().sum())
```

```python
#checking the duplicate data entry
print("Duplicate Entries : ",notNull_data.duplicated().sum())

#drop all duplicated entry
clean_data=notNull_data.drop_duplicates()
print("\n---Check whether there are duplicate data present after the clean---- ")
print("\nDuplicate data : ",clean_data.duplicated().sum())
```

# Data Set Pre Process

```python
clean_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

#creating an instance of labelencoder
labelencoder=LabelEncoder()

#assign numerical value and storing it in another
encoded_data=clean_data.iloc[0:]

encoded_data["Campaign status"]=labelencoder.fit_transform(clean_data["Campaign status"])
encoded_data["Campaign"]=labelencoder.fit_transform(clean_data["Campaign"])
encoded_data["Budget name"]=labelencoder.fit_transform(clean_data["Budget name"])
encoded_data["Currency code"]=labelencoder.fit_transform(clean_data["Currency code"])
encoded_data["Budget type"]=labelencoder.fit_transform(clean_data["Budget type"])
encoded_data["Status"]=labelencoder.fit_transform(clean_data["Status"])
encoded_data["Status reasons"]=labelencoder.fit_transform(clean_data["Status reasons"])
encoded_data["Campaign type"]=labelencoder.fit_transform(clean_data["Campaign type"])
encoded_data["Bid strategy type"]=labelencoder.fit_transform(clean_data["Bid strategy type"])
encoded_data["Campaign group"]=labelencoder.fit_transform(clean_data["Campaign group"])
```

```python
encoded_data.head()
```

```python
encoded_data.shape
```

```python
encoded_data
```

```python
encoded_data['CTR']=encoded_data['CTR'].str.replace(r'--', r'0.00')
```

```python
encoded_data['CTR'] = (pd.to_numeric(encoded_data['CTR'].str[:-1])
                            .div(100)
                            .mask(encoded_data['CTR'] == '%', 0))
```

```python
encoded_data.head(2)
```

```python
#drop the Day column
test_data=encoded_data.drop(['Day'],axis=1)

x=test_data.drop('Conv. value',axis=1)
x.head()
```

```python
x.shape
```

```python
#seperate income column as y
y=test_data['Conv. value']
print(y)
```

### Split train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print("Training Set of Feature : ",x_train.shape)
print("Test Set of Features : ",x_test.shape)
print("Training set of Target : ",y_train.shape)
print("Test set of Target : ",y_test.shape)
x_train.head()
```

## Prediction

```python
from sklearn.naive_bayes import GaussianNB
BayesModel=GaussianNB()
BayesModel.fit(x_train,y_train)
y_predictedBayes=BayesModel.predict(x_test)
print("Predicted Value by the model\n",y_predictedBayes)
print("\nTrue Value of Test Set\n",y_test)
```

```python
#accuracy of the model (naive bayes classifier)
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score
accuracy=accuracy_score(y_test,y_predictedBayes)
precision=precision_score(y_test,y_predictedBayes)
f1=f1_score(y_test,y_predictedBayes)
recall=recall_score(y_test,y_predictedBayes)
print("Accuracy of the bayes classifier is : %.2f%%"%(accuracy*100))
print("Precision of the bayes classifier is : %.2f%%"%(precision*100))
print("Recall of the bayes classifier is : %.2f%%"%(recall*100))
print("F1_Score of the bayes classifier is : %.2f%%"%(f1*100))
```

### change features into scaler and then predict

```python
from sklearn.preprocessing import StandardScaler

x0_train=StandardScaler().fit(x_train).transform(x_train)
x0_test=StandardScaler().fit(x_test).transform(x_test)
y0_train=y_train.iloc[0:]
y0_test=y_test.iloc[0:]
from sklearn.naive_bayes import GaussianNB
BayesModel2=GaussianNB()
BayesModel2.fit(x0_train,y0_train)
y0_predictedBayes=BayesModel2.predict(x0_test)
from sklearn.metrics import accuracy_score
accuracy1=accuracy_score(y0_predictedBayes,y0_test)
precision1=precision_score(y0_predictedBayes,y0_test)
f11=f1_score(y0_predictedBayes,y0_test)
recall1=recall_score(y0_predictedBayes,y0_test)
print("Accuracy of the bayes classifier is : %.2f%%"%(accuracy1*100))
print("Precision of the bayes classifier is : %.2f%%"%(precision1*100))
print("Recall of the bayes classifier is : %.2f%%"%(recall1*100))
print("F1_Score of the bayes classifier is : %.2f%%"%(f11*100))
```
