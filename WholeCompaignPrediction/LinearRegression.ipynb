# preprocessing

```python
newData=df.sort_values(by=['Day'], ascending=True)
newData.head()
```

```python
newData=newData.reset_index()
```

```python
processingData=newData.drop(['index','Day','Campaign status','Campaign','Budget name','Currency code','Budget type','Status','Status reasons','CTR','Avg. CPC','Campaign group'], axis='columns', inplace=False)
```

```python
#as type is a categorical need to transform the categorical features into numerical
from sklearn.preprocessing import LabelEncoder
import pickle

labelencoder = LabelEncoder()

# Fit the encoder on your training data (e.g., 'my_column')
labelencoder.fit(processingData['Campaign type'])

processingData['encoded_campaign_type'] = labelencoder.transform(processingData['Campaign type'])
mapping_campaign_type = dict(zip(labelencoder.classes_, labelencoder.transform(labelencoder.classes_)))

# Save it to a file (e.g., 'label_mapping.pkl')
with open('mapping_campaign_type.pkl', 'wb') as file:
    pickle.dump(mapping_campaign_type, file)

print(processingData['encoded_campaign_type'])
```

```python
le = LabelEncoder()

# Fit the encoder on your training data (e.g., 'my_column')
le.fit(processingData['Bid strategy type'])

processingData['encoded_bid_strategy_type'] =le.transform(processingData['Bid strategy type'])
mapping_bid_strategy_type = dict(zip(le.classes_, le.transform(le.classes_)))

# Save it to a file (e.g., 'label_mapping.pkl')
with open('mapping_bid_strategy_type.pkl', 'wb') as file:
    pickle.dump(mapping_bid_strategy_type, file)

print(processingData['encoded_bid_strategy_type'])
```

# linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
print("x_train: " + str(x_train.shape))
print("x_test: " + str(x_test.shape))
print("y_train: " + str(y_train.shape))
print("y_test: " + str(y_test.shape))
```

```python
LinearRegr = LinearRegression()
```

```python
LinearRegr.fit(x_train, y_train)
```

```python
print('LR Coefficients: \n', LinearRegr.coef_)
print('LR Intercept: \n', LinearRegr.intercept_)
```

```python
y_pred = LinearRegr.predict(x_test)
```

```python
result=x_test
```

```python
result['actual conversion value']=y_test
```

```python
result['predicted conversion value']=y_pred
```

```python
result
```

```python
plt.scatter(x_test['Cost'], y_test, color ='b')
plt.plot(x_test['Cost'], y_pred, color ='k')
plt.show()
```

```python
from sklearn import metrics
print("Training R-squared: ",round(metrics.r2_score(y_test,y_pred),2))
print("Training Explained Variation: ",round(metrics.explained_variance_score(y_test,y_pred),2))
print('Training MAPE:', round(metrics.mean_absolute_percentage_error(y_test,y_pred), 2)) 
print('Training Mean Squared Error:', round(metrics.mean_squared_error(y_test,y_pred), 2)) 
print("Training RMSE: ",round(np.sqrt(metrics.mean_squared_error(y_test,y_pred)),2))
print("Training MAE: ",round(metrics.mean_absolute_error(y_test,y_pred),2))
```
